```python
plt.figure(figsize=(5,5))

custom_palette = ["#FF9999", "#66B2FF"]  # Example colors: light red for one gender, light blue for another

# Plotting countplot with both Gender and EthnicGroup, using the custom palette
ax=sns.countplot(data=df, x="EthnicGroup", hue="Gender", palette=custom_palette)
plt.show()
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
print(gb)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb,annot=True)
plt.show()
```

```python
gp=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","WritingScore":"mean","ReadingScore":"mean"})
print(gp)
```

```python

plt.figure(figsize=(4,4))
sns.heatmap(gp,annot=True)
plt.title("Heatmap of parent marital status v/s students' score in different subjects ")
plt.show()
```
